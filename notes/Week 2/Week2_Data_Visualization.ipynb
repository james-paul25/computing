`Matplotlib` is a popular Python library used for creating static, animated, and interactive visualizations. It offers a variety of plots and charts, making it a powerful tool for data visualization in scientific, engineering, and financial fields.

## ***Types of Charts***

```python
# Step 1: Import matplotlib pyplot module
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Step 2: Creating a sample dataframe
df = pd.DataFrame({
    'a' : [1, 2, 3, 4, 5],
    'b' : [1, 4, 9, 16, 25]
})
```

1. `Line/Plot Chart`: A line plot or line chart is used to display data points that are connected by a line. It is particularly useful for showing trends or changes in data over time. The x-axis typically represents time intervals, while the y-axis represents the value of the variable.

```python
# line plot the data points for (x, y)
plt.plot(df['a'], df['b'])

# Setting the title of the plot
plt.title("First Line Chart")

# Labelling the x-axis
plt.xlabel('Number')

# Labelling the y-axis
plt.ylabel('Squares')

# To display the plot
plt.show()
```

2. `Bar Chart`: A bar chart uses rectangular bars to represent different categories of data, with the length or height of each bar proportional to the value it represents. It is used to compare different categories or groups. Bars can be plotted vertically or horizontally.

```python
# bar chart the data points for (x, y)
plt.bar(df['a'], df['b'], color = ["red", "green", "blue", "yellow", "pink"])

# Setting the title of the plot
plt.title("First Bar Chart")

# Labelling the x-axis
plt.xlabel('Number')

# Labelling the y-axis
plt.ylabel('Squares')

# To display the plot
plt.show()
```

3. `Pie Chart`: A pie chart is a circular chart divided into sectors, each representing a proportion of the whole. It shows the parts-to-whole relationship, making it easy to see the relative sizes of different categories. However, it can be less effective when there are many categories.

```python
# generating a pie chart 
plt.pie(df['b'], labels=df['b'], autopct='%1.1f%%', colors = ["olive", "salmon", "blue", "yellow", "violet"])

# Setting the title of the plot
plt.title("First Pie Chart")

# Adding legend
plt.legend(df['b'])

# To display the plot
plt.show()
```

4. `Scatter Plot`: A scatter plot uses Cartesian coordinates to display values for two variables for a set of data. Each point on the graph represents an individual data point with x and y values. Scatter plots are useful for identifying relationships, patterns, or correlations between two variables.

```python
# scatter plot the data points for (x, y)
plt.scatter(df['a'], df['b'], color='green')
# plt.plot(df['a'], df['b'], color='red')
# plt.bar(df['a'], df['b'], color='blue', width=0.1)

# Setting the title of the plot
plt.title("First Scatter Plot")

# Labelling the x-axis
plt.xlabel('Number')

# Labelling the y-axis
plt.ylabel('Squares')

# To display the plot
plt.show()
```

## ***Subplots***

```python
# Displaying 2 charts in one line, known as subplots
# creating 1 row with 2 columns to house the charts
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

# Line Chart
ax1.title.set_text('Line Chart')
ax1.plot(df['a'], df['b'])

# Pie Chart
ax2.title.set_text('Pie Chart')
ax2.pie(df['b'], labels=df['b'], autopct='%1.1f%%')

# # Bar Chart
ax3.title.set_text('Bar Chart')
ax3.bar(df['a'], df['b'], color = "red")

# # Scatter Plot
ax4.title.set_text('Scatter Chart')
ax4.scatter(df['a'], df['b'], color='green')

# To display the plot
plt.show()
```

## ***Adding Images to a Plot***

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Setting the title of the plot
plt.title("First Image in a Plot")

# Add a label and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Load and display an image
img = mpimg.imread('image1.jpg')
# plt.imshow(img)

# Adding an plot
plt.plot(df['a'], df['b'])

# Adjust extent and alpha as needed
plt.imshow(img, extent=[0, 50, 0, 50], alpha=0.5) 
```
